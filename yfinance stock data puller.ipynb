```python
# Developed by Ben Cooney www.linkedin.com/in/cooneyben
# Derived from https://www.youtube.com/watch?v=037_883wHGo&t=0s

import yfinance as yf
import pandas as pd
#define target stock
target = 'ADW-A.TO'
#define start date YYYY/MM/DD
startdate = '2019-01-01'
#save to filepath
filepath = "R:/" + target + startdate
```

```python
# as of version 2.55.0 the data is automatically adjusted, it is possible to get the unadjusted data with 'auto_adjust=False' as an arugment in download.
df = yf.download(target, start= startdate)
```

```python
#view data as downloaded 
df
```

```python
#remove multi index column / flatten out
df = df.xs(target, axis=1, level = 'Ticker')
```

```python
df
```

```python
#export data to .csv
df.to_excel(filepath + '.xlsx') 
```
