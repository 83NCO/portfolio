```python
'''Write a program that prompts the user for a measurement in meters and then converts it
to miles, feet, and inches.'''


x = input("enter a measurement in meters:")
# Conversion numbers taken from google conversion tool. (google "meters to miles")
to_miles_constant = 0.000621371
to_feet_constant = 3.2808388799999996799
to_inches_constant = 39.370066559999998

print(
    "your number is: " + str(int(x) * to_miles_constant) + " miles, " +
      str(int(x) * to_feet_constant) + " feet, and " +
      str(int(x) * to_inches_constant) + " inches."
      )

#Instructions didnt specify that it needed to be rounded to 2 decimals. long floats are technically more accurate ;)
```

```python
'''Write a program that reads a floating-point number and prints “zero” if the number is zero.
Otherwise, print “positive” or “negative”. Add “small” if the absolute value of the number is less
than 1,000,000 or “large” if it exceeds 1,000,000. (20%)'''

user_input = input("Enter a number: ")

if float(user_input) == 0:
    sign = "zero"
elif float(user_input) < 0:
    sign = "negative"
else:
    sign = "positive"

if abs(float(user_input)) == 0:
    size = ""
elif abs(float(user_input)) < 1000000:
    size = "small"
else:
    size = "large"

print(size,sign)
```

```python
"""Write a program that reads a set of floating-point values. Ask the user to enter the values, then print
    •the average of the values.
        •the smallest of the values.
        •the largest of the values.
        •the range, that is the difference between the smallest and largest.
        """

import numpy

ulist = []

while True:
    try:
        user_input = int(input("please enter integers, enter 0 to exit. "))
        if user_input != 0:
            ulist.append(user_input)
        else:
            break
    except ValueError:
        print("enter an integer, enter 0 to quit.")

print(ulist)
for value in ulist:
    print(value)
print(f" the average is: {numpy.average(ulist)}")
print(f" the min is: {min(ulist)}")
print(f" the max is: {max(ulist)}")
print(f" the range is: {max(ulist) - min(ulist)}")
```

```python
"""Write a program with loops that compute the sum of all odd numbers between *a* and *b*,
where a and b are inputs. (20%)"""

#input validation
while True:
    try:
        a = int(input("Enter an integer: "))
    except ValueError:
        print("enter an integer: ")
    break

while True:
    try:
        b = int(input("Enter an integer: "))
    except ValueError:
        print("enter an integer: ")
    break

#determine range, create empty list. 
range = max(a,b) - min(a,b)
group = []
# i as a counter, set to value from range. 
i = range
# add odd numbers in range to list.
while i > 0:
    if i % 2 == 1:
        group.append(i)
    i -= 1

xsum = 0
for number in group:
        xsum += number
print(group)
print(f"The sum of all odd numbers between {a} and {b} is {xsum}: ")
```

```python
'''Write programs that read a line of input as a string and print the string, with all vowels replaced
by an underscore.'''

while True:
    try:
        user_input = str(input("Enter a string: "))
    except ValueError:
        print("your input must be a string.")
    break
    

#replace cant take a list as its first argument, so ill just use a for loop. 
vlist = ["a", "e", "i", "o", "u"]

for vowel in vlist:
    user_input = user_input.replace(vowel,"_")

print(user_input)

```

```python
"""Write a program that computes the number of years required to triple the initial balance. And the
user inputs initial balance (positive), and the rate (positive)"""

while True:
    try:
        user_balance = float(input("Enter the balance: "))
        if user_balance <= 0:
            print("Error: your input must be positive, please relaunch program.")
    except ValueError:
        print("your input must be a number.")
    break

while True:
    try:
        user_rate = float(input("Enter the rate as a percentage, e.g. enter 3 for 3%: "))
        if user_rate <= 0:
            print("Error: your input must be positive, please relaunch program.")
    except ValueError:
        print("your input must be a number.")
    break
#user rate conversion
user_rate = user_rate / 100 + 1

def time_to_triple(balance, rate):
    target = balance * 3
    i = 0
    while balance <= target:
        balance = balance * rate
        i += 1
    return i

print(time_to_triple(user_balance, user_rate))
```
